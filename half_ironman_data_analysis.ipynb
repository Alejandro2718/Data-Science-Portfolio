```python
#rename columns
```

```python
# Gender                   object
# AgeGroup                 object
# AgeBand                   int64
# Country                  object
# EventYear                 int64
# EventLocation            object
# SwimTime                  int64
# Transition1Time           int64
# BikeTime                  int64
# Transition2Time           int64
# RunTime                   int64
# FinishTime                int64
# athlete_year_of_birth     int64
```

```python
df2 = df2.rename(columns = {'Gender': 'gender',
                            'AgeGroup': 'age_group',
                            'AgeBand': 'age',
                            'Country': 'athlete_country',
                            'EventYear': 'event_year',
                             'EventLocation': 'event_location',
                             'SwimTime': 'swim_time',
                             'Transition1Time': 't1_time',
                             'BikeTime': 'bike_time',
                             'Transition2Time': 't2_time',
                             'RunTime': 'run_time',
                             'FinishTime': 'finish_time',
                             'athlete_year_of_birth': 'birth_year'
                 })
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3 = df2[['age', 'gender', 'age_group', 'birth_year', 'athlete_country', 'event_location', 'event_year', 'swim_time', 't1_time', 'bike_time', 't2_time', 'run_time', 'finish_time']]
```

```python
#difference in finish time female and male
```

```python
df3.groupby(['age', 'gender'])['finish_time'].mean()
```

```python
#what country are the best and more atheltes from in the World Championship
```

```python
df3.query('event_location == " Championship"').groupby('athlete_country')['finish_time'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>100')
```

```python
#what country are the worst and more atheltes from in the World Championship
```

```python
df3.query('event_location == " Championship"').groupby('athlete_country')['finish_time'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('mean>22300 and count > 1')
```

```python
#From which country were the faster swimmers
```

```python
swim_stats = df3.groupby('athlete_country')['swim_time'].agg(['mean', 'count']).query('count >= 10').assign(mean_min=lambda x: x['mean'] / 60).sort_values('mean', ascending=True)
```

```python
sns.barplot(data=swim_stats, x='mean_min', y=swim_stats.index)
```

```python
#In conclusion, France, followed by the USA, produces better and more athletes who qualify for the World Championship.
```
